```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the CSV file
df = pd.read_csv("sample_internship_data.csv")

# Step 3: View first few rows
df.head()
# Count of applications per domain
domain_counts = df['Domain'].value_counts()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=domain_counts.index, y=domain_counts.values, palette='viridis')
plt.xticks(rotation=45)
plt.title("Applications per Internship Domain")
plt.xlabel("Internship Domain")
plt.ylabel("Number of Applications")
plt.tight_layout()
plt.show()
# Filter students who participated
participated = df[df['Participation_Status'] == 'Participated']

# Participation rate calculation
participation_rate = participated['Domain'].value_counts() / df['Domain'].value_counts() * 100

# Plot
plt.figure(figsize=(10,6))
participation_rate.sort_values(ascending=False).plot(kind='bar', color='green')
plt.title("Participation Rate (%) by Domain")
plt.ylabel("Participation Rate (%)")
plt.xlabel("Domain")
plt.tight_layout()
plt.show()
# Convert Applied_Date to datetime if not already
df['Applied_Date'] = pd.to_datetime(df['Applied_Date'])

# Group applications by month
monthly_trend = df.groupby([df['Applied_Date'].dt.to_period("M"), 'Domain']).size().unstack().fillna(0)

# Plot
monthly_trend.plot(figsize=(12,6), marker='o')
plt.title("Monthly Trend of Internship Applications by Domain")
plt.ylabel("Applications")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
